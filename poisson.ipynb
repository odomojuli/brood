```python
# Performance gains can be made through use of Cython and better indexing
import numpy as np

# For plotting
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
# lam
# float or array_like of float
# Expected number of events occurring in a fixed-time interval, must be >= 0.
# A sequence must be broadcastable over the requested size.
# In other words, the density

# From field studies, human response time averages around 274ms
# Sample across 10,000 samples
samples=10000
s = rng.poisson(lam=274,size=samples)
count, bins, ignored = plt.hist(s, 14, density=True)
plt.show()
```

```python
# These response times approximate human reaction.
# Assign each value in the sequence to a response delay value.
# Values are assigned as numpy.int64, maybe excessive memory for our purposes.
import time
samples=1000

# https://humanbenchmark.com/tests/reactiontime
# Set lam=274
s = rng.poisson(lam=274,size=samples)
times = []
# Value divided 1000 to convert from s to ms
s = s/1000
# running loop from 0 to 4
for i in range(samples):
    # printing numbers
    t0 = time.time();  # start timing
    # Time delay with samples zero-indexed
    time.sleep(s[i-1])
    t1 = time.time()
    print('Current Iteration:'+str(i)+', Elapsed time:'+str(t1 - t0)+' seconds.')
    #Convert back to seconds for reference
    times.append(1000*(t1-t0))
print(times)
plt.hist(times)

#
```

```python
# Sum initial set of samples
sumsamples = sum(s)*1000
# Sum post-process set of recorded times
sumtimes = sum(times)
# Take difference
diffsum = (sumtimes-sumsamples)
print('Sum of sample times: '+str(sumsamples/1000)+" seconds")
print('Sum of recorded times: '+str(sumtimes/1000)+" seconds")
print('Difference: '+str(diffsum/1000)+" seconds")
```
